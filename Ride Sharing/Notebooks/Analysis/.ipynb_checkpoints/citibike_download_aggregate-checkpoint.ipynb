```python
import pandas as pd
import json
from shapely.geometry import Point
import geopandas as gp
import urllib
import matplotlib.pyplot as plt
%matplotlib inline
```

# Citibike Stations

```python
url = "https://feeds.citibikenyc.com/stations/stations.json"
response = urllib.urlopen(url)
stations = json.loads(response.read())
    
# https://feeds.citibikenyc.com/stations/stations.json
```

```python
stations
```

```python
st = map(lambda x: (x['id'],x['latitude'], x['longitude'], x['stAddress1']),stations['stationBeanList'])
```

```python
cb_stations = pd.DataFrame(st, columns=['id', 'latitude', 'longitude', 'station_address'])
```

```python
cb_stations.duplicated().sum()
```

```python
cb_stations.head()
```

# Merge stations with taxi zones

```python
tzones = gp.read_file("../../Data/taxi_zones/taxi_zones_updated.shp")
```

```python
tzones.to_crs(epsg=4326, inplace=True)
```

```python
crs = {'init': 'epsg:4326', 'no_defs': True}
geometry = [Point(xy) for xy in zip(cb_stations["longitude"], cb_stations["latitude"])]
cb_stations = cb_stations.drop(["longitude", "latitude"], axis=1)
cb_stations = gp.GeoDataFrame(cb_stations, crs=crs, geometry=geometry)
```

```python
print "Spatially joining cb_stations and taxi zones"
cb_stations_geo = gp.sjoin(tzones, cb_stations) 
cb_stations_geo.drop(['borough', u'OBJECTID', 'Shape_Area', 'Shape_Leng',
                    u'geometry', 'index_right', 'zone'], axis = 1, inplace = True)  
cb_stations_geo.reset_index(inplace=True, drop=True)
```

# Get 2013 and 2014 data

```python
cb0713 = pd.read_csv("../../Data/Citibike/201307-citibike-tripdata.csv")
```

```python
cb0713 = cb0713[['starttime', 'start station id']]
cb0713["starttime"] = pd.to_datetime(cb0713["starttime"])
```

```python
cb07 = pd.merge(cb0713, cb_stations_geo, left_on='start station id', right_on= 'id', how='left')
```

```python
cb07[cb07.LocationID.isnull()]['start station id'].unique()
```

```python
cb07.dropna(inplace=True)
```

```python
cb07_agg = cb07.set_index('starttime').groupby([pd.TimeGrouper('D'),'LocationID']).agg({'start station id':'count'})
cb07_agg.reset_index('LocationID', inplace=True)
    
cb07_agg.columns = ['locationid', 'count']
```
