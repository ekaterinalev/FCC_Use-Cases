```python
station_bounds = pd.read_csv('data_raw_cb/deployment_bounds.csv')
station_bounds.drop(['Unnamed: 0'],axis=1,inplace=True)
station_bounds[station_bounds.station_id == choice]
```

```python
bufs_3 = gp.GeoDataFrame.from_file('data_raw_cb/bufs_3.shp')
bufs_3.crs = {'init': 'epsg:2263', 'no_defs': True}
bufs_3 = bufs_3.to_crs(epsg=2263)
bufs_3.crs

bufs_c = bufs_3[(bufs_3.station_id == choice) & (bufs_3.radius == radius)]
bufs_c.reset_index(drop=True, inplace = True)
bufs_c
```

```python
idxx = 562
station_bounds.iloc[idxx][2]
```

```python
def choose_col_do_2013(do_data_2013):
    do_2013 = do_data_2013[['pickup_datetime','dropoff_datetime','dropoff_longitude','dropoff_latitude','trip_distance']]
    do_2013['num_trips'] = do_data_2013['trip_distance']
    do_2013.rename(columns={'pickup_datetime':'pu_time','dropoff_datetime':'do_time',
                            'dropoff_longitude':'do_lon', 'dropoff_latitude':'do_lat'},
              inplace=True)
    return do_2013


def choose_col_do_2014(do_data_2014):
    do_2014 = do_data_2014[[' pickup_datetime',' dropoff_datetime',' dropoff_longitude',' dropoff_latitude',' trip_distance']]
    do_2014['num_trips'] = do_data_2014[' trip_distance']
    do_2014.rename(columns={' pickup_datetime':'pu_time',' dropoff_datetime':'do_time',
                            ' dropoff_longitude':'do_lon', ' dropoff_latitude':'do_lat', ' trip_distance':'trip_distance'},
              inplace=True)
    return do_2014

def choose_col_do_2015(do_data_2015):
    do_2015 = do_data_2015[['tpep_pickup_datetime','tpep_dropoff_datetime','dropoff_longitude','dropoff_latitude','trip_distance']]
    do_2015['num_trips'] = do_data_2015['trip_distance']
    do_2015.rename(columns={'tpep_pickup_datetime':'pu_time','tpep_dropoff_datetime':'do_time',
                            'dropoff_longitude':'do_lon', 'dropoff_latitude':'do_lat'},
              inplace=True)
    return do_2015

def choose_col_do_2016(do_data_2016):
    do_2016 = do_data_2016[['tpep_pickup_datetime','tpep_dropoff_datetime','dropoff_longitude','dropoff_latitude','trip_distance']]
    do_2016['num_trips'] = do_data_2016['trip_distance']
    do_2016.rename(columns={'tpep_pickup_datetime':'pu_time','tpep_dropoff_datetime':'do_time',
                            'dropoff_longitude':'do_lon', 'dropoff_latitude':'do_lat'},
              inplace=True)
    return do_2016


#######################

def trim_bounds_do(do_data):

    startTime = datetime.datetime.now()

    min_lon = station_bounds.iloc[idxx][3]
    min_lat = station_bounds.iloc[idxx][4] 
    max_lon = station_bounds.iloc[idxx][5]
    max_lat = station_bounds.iloc[idxx][6]    
    
    do_trim = do_data[(do_data['do_lon'] > min_lon) & (do_data['do_lon'] < max_lon) &
                      (do_data['do_lat'] > min_lat) & (do_data['do_lat'] < max_lat)]
    

    geometry = [Point(xy) for xy in zip(do_trim['do_lon'], do_trim['do_lat'])]
    crs = {'init': 'epsg:4326', 'no_defs': True}
#    crs = {'init': 'epsg:2263', 'no_defs': True}
    do_data_points = gp.GeoDataFrame(do_trim, crs=crs, geometry=geometry)       
    do_data_points = do_data_points.to_crs(epsg=2263)
    
    print datetime.datetime.now() - startTime, "Shrunk Dataset & Created Points"

    return do_data_points

#############################

def join_do(rides_do):
    
    startTime = datetime.datetime.now()

    mg = sjoin(rides_do, bufs_c, how='inner', op='within')
    
    print datetime.datetime.now() - startTime,"Spatial join done"

    return mg
    

    
def trim_by_date_do(do_dt):

    startTime = datetime.datetime.now()
        
    do_dt['start'] = pd.to_datetime(do_dt['pu_time'])
    do_dt['end'] = pd.to_datetime(do_dt['do_time'])
    do_dt['trip_time'] = (do_dt['end'] - do_dt['start']).dt.total_seconds()/60
    do_dt.reset_index(drop=True, inplace = True)

    print datetime.datetime.now() - startTime,"Made trip time"
    
    return do_dt

    
    
def agg_do(do_ag):
    
    startTime = datetime.datetime.now()

    do_ag['do_date'] = do_ag['end'].dt.date
    data_do_ag = do_ag[['do_date','station_id','radius','num_trips','trip_distance','trip_time']].sort(['station_id','radius','do_date'])
    data_do_ag.reset_index(inplace = True)

    data_do_ag = data_do_ag.groupby(['station_id','radius','do_date']).agg({'num_trips': "count", 'trip_distance':"mean",'trip_time':"mean"})
    
    print datetime.datetime.now() - startTime,"Aggregated by date"
    
    data_do_ag.reset_index(inplace = True)

    return data_do_ag

def run_all_13(taxi_data):
    txd = choose_col_do_2013(taxi_data)
    txd = trim_bounds_do(txd)
    txd = join_do(txd)
    txd = trim_by_date_do(txd)
    txd = agg_do(txd)
    
    return txd

def run_all_14(taxi_data):
    txd = choose_col_do_2014(taxi_data)
    txd = trim_bounds_do(txd)
    txd = join_do(txd)
    txd = trim_by_date_do(txd)
    txd = agg_do(txd)
    
    return txd

def run_all_15(taxi_data):
    txd = choose_col_do_2015(taxi_data)
    txd = trim_bounds_do(txd)
    txd = join_do(txd)
    txd = trim_by_date_do(txd)
    txd = agg_do(txd)
    
    return txd

def run_all_16(taxi_data):
    txd = choose_col_do_2016(taxi_data)
    txd = trim_bounds_do(txd)
    txd = join_do(txd)
    txd = trim_by_date_do(txd)
    txd = agg_do(txd)
    
    return txd
```
