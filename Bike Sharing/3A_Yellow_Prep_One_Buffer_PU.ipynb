```python
def choose_col_pu_2013(pu_data_2013):
    pu_2013 = pu_data_2013[['pickup_datetime','dropoff_datetime','pickup_longitude','pickup_latitude','trip_distance']]
    pu_2013['num_trips'] = pu_data_2013['trip_distance']
    pu_2013.rename(columns={'pickup_datetime':'pu_time','dropoff_datetime':'do_time',
                            'pickup_longitude':'pu_lon', 'pickup_latitude':'pu_lat'},
              inplace=True)
    return pu_2013


def choose_col_pu_2014(pu_data_2014):
    pu_2014 = pu_data_2014[[' pickup_datetime',' dropoff_datetime',' pickup_longitude',' pickup_latitude',' trip_distance']]
    pu_2014['num_trips'] = pu_data_2014[' trip_distance']
    pu_2014.rename(columns={' pickup_datetime':'pu_time',' dropoff_datetime':'do_time',
                            ' pickup_longitude':'pu_lon', ' pickup_latitude':'pu_lat', ' trip_distance':'trip_distance'},
              inplace=True)
    return pu_2014

def choose_col_pu_2015(pu_data_2015):
    pu_2015 = pu_data_2015[['tpep_pickup_datetime','tpep_dropoff_datetime','pickup_longitude','pickup_latitude','trip_distance']]
    pu_2015['num_trips'] = pu_data_2015['trip_distance']
    pu_2015.rename(columns={'tpep_pickup_datetime':'pu_time','tpep_dropoff_datetime':'do_time',
                            'pickup_longitude':'pu_lon', 'pickup_latitude':'pu_lat'},
              inplace=True)
    return pu_2015

def choose_col_pu_2016(pu_data_2016):
    pu_2016 = pu_data_2016[['tpep_pickup_datetime','tpep_dropoff_datetime','pickup_longitude','pickup_latitude','trip_distance']]
    pu_2016['num_trips'] = pu_data_2016['trip_distance']
    pu_2016.rename(columns={'tpep_pickup_datetime':'pu_time','tpep_dropoff_datetime':'do_time',
                            'pickup_longitude':'pu_lon', 'pickup_latitude':'pu_lat'},
              inplace=True)
    return pu_2016


#######################

def trim_bounds_pu(pu_data):
    
    #min_lon = station_bounds[station_bounds['station_id'] == bounds]['min_lon'][0]
    #max_lon = station_bounds[station_bounds['station_id'] == bounds]['max_lon'][0]
    #min_lat = station_bounds[station_bounds['station_id'] == bounds]['min_lat'][0]
    #max_lat = station_bounds[station_bounds['station_id'] == bounds]['max_lat'][0]

    startTime = datetime.datetime.now()

    #b = 430
    
    min_lon = station_bounds.iloc[idxx][3]
    min_lat = station_bounds.iloc[idxx][4] 
    max_lon = station_bounds.iloc[idxx][5]
    max_lat = station_bounds.iloc[idxx][6]    
    
    pu_trim = pu_data[(pu_data['pu_lon'] > min_lon) & (pu_data['pu_lon'] < max_lon) &
                      (pu_data['pu_lat'] > min_lat) & (pu_data['pu_lat'] < max_lat)]
    

    geometry = [Point(xy) for xy in zip(pu_trim['pu_lon'], pu_trim['pu_lat'])]
    crs = {'init': 'epsg:4326', 'no_defs': True}
#    crs = {'init': 'epsg:2263', 'no_defs': True}
    pu_data_points = gp.GeoDataFrame(pu_trim, crs=crs, geometry=geometry)       
    pu_data_points = pu_data_points.to_crs(epsg=2263)
    
    print datetime.datetime.now() - startTime, "Shrunk Dataset & Created Points"

    return pu_data_points

#############################

def join_pu(rides_pu):
    
    startTime = datetime.datetime.now()

    mg = sjoin(rides_pu, bufs_c, how='inner', op='within')
    
    print datetime.datetime.now() - startTime,"Spatial join done"

    return mg
    

    
def trim_by_date_pu(pu_dt):

    startTime = datetime.datetime.now()
        
    pu_dt['start'] = pd.to_datetime(pu_dt['pu_time'])
    pu_dt['end'] = pd.to_datetime(pu_dt['do_time'])
    pu_dt['trip_time'] = (pu_dt['end'] - pu_dt['start']).dt.total_seconds()/60
    pu_dt.reset_index(drop=True, inplace = True)

    print datetime.datetime.now() - startTime,"Made trip time"
    
    return pu_dt

    
    
def agg_pu(pu_ag):
    
    startTime = datetime.datetime.now()

    pu_ag['pu_date'] = pu_ag['start'].dt.date
    data_pu_ag = pu_ag[['pu_date','station_id','radius','num_trips','trip_distance','trip_time']].sort(['station_id','radius','pu_date'])
    data_pu_ag.reset_index(inplace = True)

    data_pu_ag = data_pu_ag.groupby(['station_id','radius','pu_date']).agg({'num_trips': "count", 'trip_distance':"mean",'trip_time':"mean"})
    
    print datetime.datetime.now() - startTime,"Aggregated by date"
    
    data_pu_ag.reset_index(inplace = True)

    return data_pu_ag

def run_all_13(taxi_data):
    txd = choose_col_pu_2013(taxi_data)
    txd = trim_bounds_pu(txd)
    txd = join_pu(txd)
    txd = trim_by_date_pu(txd)
    txd = agg_pu(txd)
    
    return txd

def run_all_14(taxi_data):
    txd = choose_col_pu_2014(taxi_data)
    txd = trim_bounds_pu(txd)
    txd = join_pu(txd)
    txd = trim_by_date_pu(txd)
    txd = agg_pu(txd)
    
    return txd

def run_all_15(taxi_data):
    txd = choose_col_pu_2015(taxi_data)
    txd = trim_bounds_pu(txd)
    txd = join_pu(txd)
    txd = trim_by_date_pu(txd)
    txd = agg_pu(txd)
    
    return txd

def run_all_16(taxi_data):
    txd = choose_col_pu_2016(taxi_data)
    txd = trim_bounds_pu(txd)
    txd = join_pu(txd)
    txd = trim_by_date_pu(txd)
    txd = agg_pu(txd)
    
    return txd
```
